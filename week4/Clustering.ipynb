```python
import numpy as np
from matplotlib import pyplot as plt

X = np.hstack([np.arange(8) * np.array([[1], [0]]), 
               np.arange(8) * np.array([[1], [0]]) + np.array([[0],[7]])])
X = X.T
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2)
# KMeans?
kvec = km.fit_predict(X)
kvec

plt.scatter(X[kvec==0,0], X[kvec==0,1])
plt.scatter(X[kvec==1,0], X[kvec==1,1])
plt.ylim([0,7])
plt.show()
```
